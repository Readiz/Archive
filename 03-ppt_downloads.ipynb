```python
# your code here
from lxml import html
from urllib.request import urlopen

page_open = urlopen('file:../data/db.html')
page = html.parse(page_open)
```

---
Let your automated `browser` load the course site. And __find__ all urls for ppt downloads, and __store__ it to a list `urls.` All PPT files are in the 'Course Plan' table.   

To find proper urls, you need to look up **debugger** in your web browser.
There should be **16** .ppt files found and downloaded.

```python
urls = []

### YOUR CODE HERE
root = page.getroot()
atags = root.xpath('//a') # .xpath('//a[.="PPT" or .="PPT2"]')

for a in atags:
    if (a.text != None and a.text.startswith('PPT')):
        urls.append(a.attrib.get('href'))
print(urls)
```
